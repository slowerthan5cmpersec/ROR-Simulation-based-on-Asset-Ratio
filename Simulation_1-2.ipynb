```python
import pandas as pd
import os
import natsort

sheet_id = ""
sheet_name = ""
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
df = pd.read_csv(url)
Asset_Names = list(df.columns)

df_path = "df/"

Data_List = []
Name_List = []

for i ,name in enumerate(Asset_Names):
  df_list = natsort.natsorted(os.listdir(df_path + name))
  Data_List.append([])
  Name_List.append([])

  for csv in df_list:
    try:
      df = pd.read_csv(df_path + name + '/' + csv, encoding='utf-8-sig')
      df["Date"] = pd.to_datetime(df["Date"])
      df = df.set_index("Date")
      Data_List[i].append(df)
      Name_List[i].append(csv)
    except:
      print(csv)
```

```python
Ratio_List = []

for i in range(1, 11):
  sheet_id = ""
  sheet_name = "sheet" + str(i)
  url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

  rl = pd.read_csv(url)
  Ratio_List.append(rl.iloc[0].tolist())
```

```python
import random

time_series = pd.read_csv('')
time_series['Date'] = pd.to_datetime(time_series["Date"])
time_series = time_series['Date'].tolist()
year = time_series[0].year

def date_index(n = 5):
  '''n=5: 한달에 몇 개 일자, 최대 5 (넘으면 휴장일 때문에 버그 발생)'''
  if n > 5:
    raise Exception(f"n값 5 넘기면 휴장일 때문에 버그, input n: {n}")

  time_series = pd.read_csv('')
  time_series['Date'] = pd.to_datetime(time_series["Date"])
  time_series = time_series['Date'].tolist()
  year_range = time_series[-1].year - time_series[0].year
  year = time_series[0].year

  li = []
  for i in range(year_range + 1):
    for month in range(1, 13):
      month = str(month)
      for howmany in range(n):
        stack = 0
        k = 1
        if month==2 and howmany != 0:
          k=-1
        while True:
          rn = str(random.randint(30//n*howmany+k, 30//n*(howmany+1)))
          try:
            x = pd.Timestamp(str(year+i)+"-"+month+"-"+rn)
          except:
            continue
          if pd.Timestamp(str(year+i)+"-"+month+"-"+rn) in time_series:
            break

        li.append(pd.Timestamp(str(year+i)+"-"+str(month)+"-"+rn))

  return li
```
