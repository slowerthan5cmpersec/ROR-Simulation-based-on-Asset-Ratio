## test code
> 확률통계론_프로젝트/df/z_test 에 뭐 저장됐는지 확인


```python
# test code

import yfinance as yf
import time
from datetime import datetime, timezone, timedelta

df = yf.download('005930.KS',time.strftime("%Y-%m-%d"))
df = df[['Low', 'Open', 'Close', 'High', 'Volume']]
df.columns.names = [None, None]
df = df.droplevel(level=1, axis=1)
utc_now = datetime.now(timezone.utc)
utc_kst = utc_now.astimezone(tz=timezone(timedelta(hours=9)))
df.to_csv("./drive/MyDrive/확률통계론_프로젝트/df/z_test/삼성전자_test_{0}.csv".format(str(utc_kst.strftime("%y%m%d_%H:%M:%S"))), encoding='utf-8-sig')
df.head()
```
